```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Locations = ['alley','row','midway']

DwarfDrainageData = [[0,0,0],
[86,45,78],
[86,45,78],
[86,45,103],
[86,45,103],
[173,270,216],
[189,426,235],
[223,643,272],
[367,1037,466],
[372,1325,478],
[372,1325,478],
[372,1327,481],
[372,1327,481],
[372,1381,485],
[387,1757,557],
[734,2144,935],
[996,2535,1309],
[1067,2905,1386],
[1073,3173,1391],
[1075,3175,1396],
[1076,3175,1397]]

TallDrainageData = [0,42,62,81,113,243,268,333,600,656,656,656,656,656,807,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]

Dates = pd.to_datetime(['2008-10-01',
'2009-01-21',
'2009-02-27',
'2009-03-20',
'2009-05-27',
'2009-07-24',
'2009-08-19',
'2009-09-18',
'2009-10-22',
'2009-11-19',
'2009-12-24',
'2010-01-22',
'2010-03-09',
'2010-05-10',
'2010-06-21',
'2010-08-06',
'2010-09-23',
'2010-10-21',
'2010-12-10',
'2011-01-30',
'2011-03-08']
)
DwarfDrainage = pd.DataFrame(index=Dates, columns= Locations, data = DwarfDrainageData)
TallDrainage = pd.DataFrame(index=Dates,columns = ['row'],data= TallDrainageData)
Drainage=pd.concat([DwarfDrainage,TallDrainage],keys=['Dwarf','Tall'],names=['Crop'],sort=True)
```

```python
Fig = plt.figure(figsize=(10,5))
for l in Drainage.columns:
    plt.plot(Drainage.loc['Dwarf',l],'-o',label='Dwarf ' +l)
plt.plot(Drainage.loc['Tall','row'],'-o',label='Tall row')
plt.legend()
```
