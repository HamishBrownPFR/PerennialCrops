```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
VWC = pd.read_pickle('SWC.pkl')
EP = pd.read_pickle('Transpiration.pkl')
Drainage = pd.read_pickle('Drainage.pkl')
RadInt = pd.read_pickle('RadInt.pkl')
```

```python
Data = pd.concat([VWC,EP,Drainage,RadInt],sort=True)
Data.sort_index(inplace=True)
```

```python
Fig = plt.figure(figsize=(10,15))
Ax = Fig.add_subplot(4,1,1)
plt.plot(Data.loc['Dwarf','EP(mm)'],'o',color='pink',label='PinkLady')
plt.plot(Data.loc['Tall','EP(mm)'],'o',color='purple',label='RoyalGala')
dates = Data.index.get_level_values(1)
plt.xlim(dates.min(),dates.max())
plt.ylabel('Transpiration (mm)')
Ax = Fig.add_subplot(4,1,2)
ms=4
alp = 0.2
for layer in ['SW(1)','SW(2)','SW(3)','SW(4)','SW(5)','SW(6)']:
    plt.plot(Data.loc['Dwarf',layer],'o',
             color='pink',label = 'PinkLady '+layer,ms=ms,alpha=alp,mec='k',mew=1)
    ms+=1
    alp+=0.1
ms=4
alp = 0.2
for layer in ['SW(1)','SW(2)','SW(3)','SW(4)']:
    plt.plot(Data.loc['Tall',layer],'o',
             color='purple', label = 'RoyalGala '+layer,ms=ms,alpha=alp,mec='k',mew=1)
    ms+=1.5
    alp+=0.2
plt.xlim(dates.min(),dates.max())
plt.ylabel('VSW (mm/mm)')
plt.legend()
Ax = Fig.add_subplot(4,1,3)
markers =['o','s','^']
pos=0
for location in ('alley','midway','row'):
    plt.plot(Data.loc['Dwarf',location],markers[pos],label='PinkLady '+location,color='pink',mec='k',ms=10)
    plt.plot(Data.loc['Tall',location],markers[pos],label='RoyalGala '+location,color='purple',mec='k',ms=10)
    pos+=1
plt.xlim(dates.min(),dates.max())
plt.ylabel('Accumulated drainage (mm)')
plt.legend()
Ax = Fig.add_subplot(4,1,4)
plt.plot(Data.loc['Dwarf','RadInt'],'o',color='pink',label='PinkLady',mec='k')
plt.plot(Data.loc['Tall','RadInt'],'o',color='purple',label='RoyalGala',mec='k')
plt.legend()
plt.xlim(dates.min(),dates.max())
plt.ylabel('Radiation interception')
```
