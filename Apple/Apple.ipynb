```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
VWC = pd.read_pickle('SWC.pkl')
EP = pd.read_pickle('Transpiration.pkl')
Drainage = pd.read_pickle('Drainage.pkl')
RadInt = pd.read_pickle('RadInt.pkl')
MergedIndex = VWC.index|EP.index|Drainage.index|RadInt.index
```

```python
PeachData = VWC.reindex(MergedIndex).join(EP).join(Drainage).join(RadInt)
```

```python
Fig = plt.figure(figsize=(10,15))
Ax = Fig.add_subplot(4,1,1)
plt.plot(PeachData.loc[:,'EP(mm)'],'o')
plt.xlim(PeachData.index.min(),PeachData.index.max())
plt.ylabel('Transpiration (mm)')
Ax = Fig.add_subplot(4,1,2)
for layer in ['WC0-30','WC40-70','WC80-110']:
    plt.plot(PeachData.loc[:,layer],'-',label = layer)
plt.xlim(PeachData.index.min(),PeachData.index.max())
plt.ylabel('VWC (mm/mm)')
plt.legend()
Ax = Fig.add_subplot(4,1,3)
for location in ('alley','midway','row'):
    plt.plot(PeachData.loc[:,location],'o',label=location)
plt.xlim(PeachData.index.min(),PeachData.index.max())
plt.ylabel('Accumulated drainage (mm)')
plt.legend()
Ax = Fig.add_subplot(4,1,4)
plt.plot(PeachData.loc[:,'RadInt'],'o',color='orange')
plt.xlim(PeachData.index.min(),PeachData.index.max())
plt.ylabel('Radiation interception')
```
