```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Data = pd.read_excel('C:\GitHubRepos\PerennialCrops\Peach\SWC.xlsx')    
Data.dropna(inplace=True)
```

```python
Data.loc[:,'Date'] = [pd.to_datetime(str(Data.loc[x,'day'].date()) + ' ' 
                          + str(datetime.datetime.strptime(str(int(Data.loc[x,'Time'])).zfill(4),'%H%M').time()))
                          for x in Data.index]
```

```python
Data.set_index('Date',inplace=True)
Data.drop(columns=['day','Time'],inplace=True)
```

```python
DailyData = Data.loc[Data.index.hour==23,:].resample('d').mean()
```

```python
DailyData.plot(figsize=(10,5),style='o')
```

```python
filter = ((Data.index.month==8)|(Data.index.month==7)|(Data.index.month==8))&(Data.index.year==2008)
plt.plot(Data.loc[filter,:],'o')
```
